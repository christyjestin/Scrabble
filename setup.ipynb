```python
class Tile:
    def __init__(self, letter, score, id):
        self.score = score
        self.letter = letter
        self.id = id
    def __repr__(self):
        return '(' + str(self.letter) + ' : ' + str(self.score) + ')'
```

```python
class Player:
    def __init__(self, name, position):
        self.name = name
        self.score = 0
        self.position = position
        self.rack = list()
        
    def __repr__(self):
        return '(' + str(self.position) + ', ' + str(self.name) + ', score: ' + str(self.score) + ')' 
    
    def draw(self, tiles, board):
        for tile in tiles:
            self.rack.remove(tile)
        board.pool.extend(tiles)
        shuffle(board.pool)
        self.refill(board)
    
    def refill(self, board):
        while len(self.rack)<7:
            if board.pool:
                self.rack.append(board.pool[0])
                board.pool.pop(0)
            else:
                break
                
    def show_rack(self):
        print(self.rack)
```

```python
board_values = [0,200,300,2000,3000]
class Board:
    def __init__(self, num_players):
        self.num_players = num_players
        self.pool = list()
        self.id_dict = dict()
        id = 1
        for key in distributions:
            for i in range(distributions[key]):
                self.pool.append(Tile(key, scores[key],id))
                shuffle(self.pool)
                self.id_dict[id] = Tile(key, scores[key],id)
                id+=1
        for id in board_values:
            self.id_dict[id]= Tile(' ',None,id)
        self.players = dict()
        for i in range(1, num_players+1):
            name = input('What is Player '+str(i)+"'s name?\n")
            self.players[name] = Player(name,i)
        self.board = np.load('board.npy')
        self.board_positions = list()
        self.rows = set()
        self.cols = set()
        self.row_spots = defaultdict(set)
        self.col_spots = defaultdict(set)
```

```python
word_lengths = defaultdict(set)
true_lengths = defaultdict(set)
for word in words:
    for length in range(1,22):
        if len(word)>=length:
            word_lengths[length].add(word[:length].upper())
        if len(word)==length:
            true_lengths[length].add(word.upper())
```

```python
directions_dict = dict()
def step_down(i,j):
    return (i+1,j)
def step_right(i,j):
    return (i,j+1)
def step_left(i,j):
    return (i,j-1)
def step_up(i,j):
    return (i-1,j)
directions_dict[0] = step_up
directions_dict[1] = step_right
directions_dict[2] = step_down
directions_dict[3] = step_left
reverse_dict = dict()
reverse_dict[0] = 2
reverse_dict[1] = 3
reverse_dict[2] = 0
reverse_dict[3] = 1
```

```python
def tiler(tiles, numbers):
    return_list = list()
    for i in numbers:
        return_list.append(tiles[int(i)])
    return return_list
def read_row(read_board, row):
    row_list = list()
    for i in range(15):
        row_list.append(read_board.id_dict[read_board.board[row, i]].letter)
    return row_list


def read_col(read_board, col):
    col_list = list()
    for i in range(15):
        col_list.append(read_board.id_dict[read_board.board[i,col]].letter)
    return col_list    


def read_row_letters(row_board, strip, tiles, seq, row, spot):
    tiles = tiler(tiles, seq)
    for i in range(len(tiles)):
        tiles[i] = tiles[i].letter
    attempt = list()
    j = 0
    for i in range(15):
        if tiles and strip[i]==' ' and i>=spot:
            attempt.append(tiles[0])
            tiles.pop(0)
            if not tiles:
                j = i
                tile = attempt[i]
        else:
            attempt.append(strip[i])
    strip_attempt = ''.join(attempt)
    words = strip_attempt.split()
    n = 0
    for word in words:
        if len(word)>1 and word not in word_lengths[len(word)]:
            return 0
        if len(word)>1 and word in true_lengths[len(word)]:
            n+=1
    if j==14:
        return 2
    #checks perpendicular col
    j_col = read_col(row_board, j)
    j_col[row] = tile
    col_attempt = ''.join(j_col)
    col_words = col_attempt.split()
    for word in col_words:
        if len(word)>1 and word not in true_lengths[len(word)]:
            return 0
    q = n/len(words)
    if q==0:
        q+=0.5
    if q==1:
        if not ''.join(strip[spot:j+1]).split():
            q-=0.5
    if j==14:
        q+=1
    return q
```

```python

def read_col_letters(col_board, strip, tiles, seq, col, spot):
    tiles = tiler(tiles, seq)
    for i in range(len(tiles)):
        tiles[i] = tiles[i].letter
    attempt = list()
    j = 0
    for i in range(15):
        if tiles and strip[i]==' ' and i>=spot:
            attempt.append(tiles[0])
            tiles.pop(0)
            if not tiles:
                j = i
                tile = attempt[i]
        else:
            attempt.append(strip[i])
    strip_attempt = ''.join(attempt)
    words = strip_attempt.split()
    n = 0
    for word in words:
        if len(word)>1 and word not in word_lengths[len(word)]:
            return 0
        if len(word)>1 and word in true_lengths[len(word)]:
            n+=1
    #checks perpendicular row
    j_row = read_row(col_board, j)
    j_row[col] = tile
    row_attempt = ''.join(j_row)
    row_words = row_attempt.split()
    for word in row_words:
        if len(word)>1 and word not in true_lengths[len(word)]:
            return 0
    q = n/len(words)
    if q==0:
        q+=0.5
    if q==1:
        if not ''.join(strip[spot:j+1]).split():
            q-=0.5
    if j==14:
        q+=1
    return q
```

```python

def add_letter(tiles, stem, stops):
    poss = tiles.copy()
    for i in stem:
        poss.remove(i)
    for i in stops[stem]:
        poss.remove(i)
    if poss:
        stem+=poss[0]
        return stem
    else:
        return False

def find_different(tiles, stem, stops):
    poss = tiles.copy()
    for i in stem:
        poss.remove(i)
    for i in stops[stem]:
        poss.remove(i)
    if poss:
        stem+=poss[0]
        return stem
    else:
        return False
```

```python
seven = [str(i) for i in range(7)]
def find_moves(find_board, tiles):
    moves = list()
    for row in find_board.rows:
        strip = read_row(find_board, row)
        for spot in find_board.row_spots[row]:
            stops = defaultdict(list)
            add = '0'
            truth = read_row_letters(find_board, strip, tiles, add, row, spot)
            while len(stops[''])!=7:
                if truth==0:
                    stem = add[:len(add)-1]
                    stop = add[-1]
                    stops[stem].append(stop)
                    if len(stops[''])==7: break
                    while not find_different(seven, stem, stops):
                        stop = stem[-1]
                        stem = stem[:len(stem)-1]
                        stops[stem].append(stop)
                        if len(stops[''])==7: break
                    if len(stops[''])==7: break
                    add = find_different(seven, stem, stops)
                    truth = read_row_letters(find_board, strip, tiles, add, row, spot)
                elif truth>1:
                    if truth==2:
                        moves.append((add, (row, spot,1)))
                    stem = add[:len(add)-1]
                    stop = add[-1]
                    stops[stem].append(stop)
                    if len(stops[''])==7: break
                    while not find_different(seven, stem, stops):
                        stop = stem[-1]
                        stem = stem[:len(stem)-1]
                        stops[stem].append(stop)
                        if len(stops[''])==7: break
                    if len(stops[''])==7: break
                    add = find_different(seven, stem, stops)
                    truth = read_row_letters(find_board, strip, tiles, add, row, spot)
                else:
                    if truth==1:
                        moves.append((add, (row, spot,1)))
                    stem = add
                    while not add_letter(seven, stem, stops):
                            stop = stem[-1]
                            stem = stem[:len(stem)-1]
                            stops[stem].append(stop)
                            if len(stops[''])==7: break
                    if len(stops[''])==7: break
                    add = add_letter(seven, stem, stops)
                    truth = read_row_letters(find_board, strip, tiles, add, row, spot)
             
    for col in find_board.cols:
        strip = read_col(find_board, col)
        for spot in find_board.col_spots[col]:
            stops = defaultdict(list)
            add = '0'
            truth = read_col_letters(find_board, strip, tiles, add, col, spot)
            while len(stops[''])!=7:
                if truth==0:
                    stem = add[:len(add)-1]
                    stop = add[-1]
                    stops[stem].append(stop)
                    if len(stops[''])==7: break
                    while not find_different(seven, stem, stops):
                        stop = stem[-1]
                        stem = stem[:len(stem)-1]
                        stops[stem].append(stop)
                        if len(stops[''])==7: break
                    if len(stops[''])==7: break
                    add = find_different(seven, stem, stops)
                    truth = read_col_letters(find_board, strip, tiles, add, col, spot)
                elif truth>1:
                    if truth==2:
                        moves.append((add, (spot, col,2)))
                    stem = add[:len(add)-1]
                    stop = add[-1]
                    stops[stem].append(stop)
                    if len(stops[''])==7: break
                    while not find_different(seven, stem, stops):
                        stop = stem[-1]
                        stem = stem[:len(stem)-1]
                        stops[stem].append(stop)
                        if len(stops[''])==7: break
                    if len(stops[''])==7: break
                    add = find_different(seven, stem, stops)
                    truth = read_col_letters(find_board, strip, tiles, add, col, spot)
                else:
                    if truth==1:
                        moves.append((add, (spot, col,2)))
                    stem = add
                    while not add_letter(seven, stem, stops):
                            stop = stem[-1]
                            stem = stem[:len(stem)-1]
                            stops[stem].append(stop)
                            if len(stops[''])==7: break
                    if len(stops[''])==7: break
                    add = add_letter(seven, stem, stops)
                    truth = read_col_letters(find_board, strip, tiles, add, col, spot)
    return moves
```

```python
def check_border(add_board, opening, length):
    m,n,direction = opening
    while length:
        q = add_board.board[m,n]
        if q in board_values:
            length-=1
        m,n = directions_dict[direction](m,n)
    if m not in range(0,15) or n not in range(0,15):
        return 1
```

```python
def add_cross_check(check_board, position, direction):
    i,j,bonus = position
    a,b = i,j
    p = directions_dict[reverse_dict[direction]]
    i,j = p(i,j)
    while check_board.board[i,j] not in board_values:
        i,j = p(i,j)
        if i not in range(0,15) or j not in range(0,15):
            break
    i,j = directions_dict[direction](i,j)
    d = directions_dict[direction]
    score = 0
    counter = 0
    score_multiply = 1
    while check_board.board[i,j] not in board_values:
        n = check_board.board[i,j]
        if i==a and j==b:
            if bonus == 0:
                score+=check_board.id_dict[n].score
            if bonus == 2000:
                score+=check_board.id_dict[n].score
                score_multiply *= 2
            if bonus == 3000:
                score+=check_board.id_dict[n].score
                score_multiply *= 3
            if bonus == 200:
                score+=check_board.id_dict[n].score*2
            if bonus == 300:
                score+=check_board.id_dict[n].score*3
        else:
            score+=check_board.id_dict[n].score
        i,j = d(i,j)
        counter+=1
        if i not in range(0,15) or j not in range(0,15):
            break        
    score*=score_multiply
    if counter==1:
        return 0
    else:
        return score
def calc_cross_check(check_board, position, direction):
    i,j,tile = position
    a,b = i,j
    p = directions_dict[reverse_dict[direction]]
    i,j = p(i,j)
    while check_board.board[i,j] not in board_values:
        i,j = p(i,j)
        if i not in range(0,15) or j not in range(0,15):
            break
    i,j = directions_dict[direction](i,j)
    d = directions_dict[direction]
    score = 0
    counter = 0
    score_multiply = 1
    while i!=a or j!=b:
        n = check_board.board[i,j]
        score+=check_board.id_dict[n].score
        i,j = d(i,j)
        counter+=1
    n = check_board.board[i,j]
    if n == 0:
        score+=tile.score
    if n == 2000:
        score+=tile.score
        score_multiply *= 2
    if n == 3000:
        score+=tile.score
        score_multiply *= 3
    if n == 200:
        score+=tile.score*2
    if n == 300:
        score+=tile.score*3
    i,j = d(i,j)
    counter+=1
    if i in range(0,15) and j in range(0,15):
        while check_board.board[i,j] not in board_values:
            n = check_board.board[i,j]
            score+=check_board.id_dict[n].score
            counter+=1
            if i not in range(0,15) or j not in range(0,15):
                break        
    score*=score_multiply
    if counter==1:
        return 0
    else:
        return score
```

```python
#note that add_word takes in player name as string and not a player class object
def add_word(player, add_board, opening, tiles):
    '''
    input player as string, sp as tuple, direction as 0, 1, 2, 3, tiles as list of indices for tile objects in rack list
    '''
    if check_border(add_board,opening,len(tiles))==1:
        return 'You\'re trying to play a tile outside the board!'
    for i in tiles:
        add_board.players[player].rack.remove(i)
    bingo = 0 
    if len(tiles)==7:    bingo = 50
    score = 0
    score_multiply = 1
    i,j,direction = opening
    p = directions_dict[reverse_dict[direction]]
    i,j = p(i,j)
    while add_board.board[i,j] not in board_values:
        i,j = p(i,j)
        if i not in range(0,15) or j not in range(0,15):
            break
    i,j = directions_dict[direction](i,j)
    positions = list()
    while tiles:
        n = add_board.board[i,j]
        if n not in board_values:
            score+=add_board.id_dict[n].score
        else:
            positions.append((i,j,n))
            if n == 0:
                score+=tiles[0].score
            if n == 2000:
                score+=tiles[0].score
                score_multiply *= 2
            if n == 3000:
                score+=tiles[0].score
                score_multiply *= 3
            if n == 200:
                score+=tiles[0].score*2
            if n == 300:
                score+=tiles[0].score*3
            add_board.board_positions.append((i,j))
            add_board.board[i,j] = tiles[0].id
            add_board.rows.add(i)
            add_board.cols.add(j)
            add_board.row_spots[i].update(range(max(0,j-7),min(j+2,15)))
            add_board.col_spots[j].update(range(max(0,i-7),min(i+2,15)))
            tiles.pop(0)
        i,j = directions_dict[direction](i,j)
    while add_board.board[i,j] not in board_values:
        n = add_board.board[i,j]
        score+=add_board.id_dict[n].score
        i,j = directions_dict[direction](i,j)
        if i not in range(0,15) or j not in range(0,15):
            break
    score*=score_multiply
    score+=bingo
    for position in positions:
        score+=add_cross_check(add_board,position,3-direction)
    add_board.players[player].score += score
    add_board.players[player].refill(add_board)
    for i,j in add_board.board_positions:
        add_board.row_spots[i].discard(j)
        add_board.col_spots[j].discard(i)
```

```python
def calc_word(add_board, opening, tiles):
    bingo = 0 
    if len(tiles)==7:    bingo = 50
    score = 0
    score_multiply = 1
    i,j,direction = opening
    p = directions_dict[reverse_dict[direction]]
    positions = list()
    i,j = p(i,j)
    while add_board.board[i,j] not in board_values:
        i,j = p(i,j)
        if i not in range(0,15) or j not in range(0,15):
            break
    i,j = directions_dict[direction](i,j)
    while tiles:
        n = add_board.board[i,j]
        if n not in board_values:
            score+=add_board.id_dict[n].score
        else:
            positions.append((i,j,tiles[0]))
            if n == 0:
                score+=tiles[0].score
            if n == 2000:
                score+=tiles[0].score
                score_multiply *= 2
            if n == 3000:
                score+=tiles[0].score
                score_multiply *= 3
            if n == 200:
                score+=tiles[0].score*2
            if n == 300:
                score+=tiles[0].score*3
            tiles.pop(0)
        i,j = directions_dict[direction](i,j)
    if i in range(0,15) and j in range(0,15):
        while add_board.board[i,j] not in board_values:
            n = add_board.board[i,j]
            score+=add_board.id_dict[n].score
            i,j = directions_dict[direction](i,j)
            if i not in range(0,15) or j not in range(0,15):
                break
    score*=score_multiply
    score+=bingo
    for position in positions:
        score+=calc_cross_check(add_board,position,3-direction)
    return score
```

```python
def compmove(playerboard, player):
    clear_output()
    player.show_rack()
    moves = find_moves(playerboard,player.rack)
    if moves:
        move_scores = np.zeros((len(moves)))
        for i in range(len(moves)):
            add, opening = moves[i]
            move_scores[i] = calc_word(playerboard,opening,tiler(player.rack,add))
        opt_add, opt_opening = moves[np.argmax(move_scores)]
        add_word(player.name, playerboard, opt_opening, tiler(player.rack,opt_add))
    else:
        player.draw(player.rack,playerboard)
        player.show_rack()
    print(list(playerboard.players.values()))
    

def playermove(playerboard,player):
    clear_output()
    player.show_rack()
    draw = input('Would you like to draw instead? Press Y/N: ')
    if draw == 'Y' or draw == 'y':
        order = input('Enter the placements of the tiles you want to replace. They are numbered 1 to 7 from left to right: ')
        tiles = list()
        for i in order:
            tiles.append(player.rack[int(i)-1])
        player.draw(tiles,playerboard)
        clear_output()
        player.show_rack()
    else:
        row = input('Which row would you like to start in? The rows are 1 to 15 from top to bottom: ')
        col = input('Which column would you like to start in? The columns are 1 to 15 from left to right: ')
        direction = input('Which direction would you like to go? Right is 1, down is 2: ')
        opening = (int(row)-1, int(col)-1, int(direction))
        order = input('How do you want to order the tiles? If you\'re using a tile on the board, ignore it and enter the next tile on your own rack. Enter the order as a series of numbers from 1 to 7 with no spaces: ')
        tiles = list()
        for i in order:
            tiles.append(player.rack[int(i)-1])
        add_word(player.name, playerboard, opening, tiles)
    print(list(playerboard.players.values()))
```
